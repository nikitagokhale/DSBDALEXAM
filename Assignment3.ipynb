```python
import pandas as pd
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
missing_values = df.isnull().sum()
print("Missing values:\n", missing_values)
```

```python
inconsistent_values_math = df[(df['math score'] < 0) | (df['math score'] > 100)]
inconsistent_values_reading = df[(df['reading score'] < 0) | (df['reading score'] > 100)]
inconsistent_values_writing = df[(df['writing score'] < 0) | (df['writing score'] > 100)]
print("Inconsistent values:\n",inconsistent_values_math,"\n",inconsistent_values_reading,"\n",inconsistent_values_writing )
```

```python
df_cleaned = df.dropna()
df_cleaned
```

```python
df_filled_mean = df.fillna(df.mean())
df_filled_mean
```

```python
df_filled_mode = df.fillna(df.mode().iloc[0])
df_filled_mode
```

```python
df_filled_median = df.fillna(df.median())
df_filled_median
```

```python
from sklearn.preprocessing import StandardScaler

# Assuming you want to scale the 'age' column
scaler = StandardScaler()
df = scaler.fit_transform(df)
```

```python
import numpy as np

df['math_score_log'] = np.log1p(df['math score'])
df
```

```python
df['math_score_sqrt'] = np.sqrt(df['math score'])
df
```
