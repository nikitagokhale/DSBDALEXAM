```python
import pandas as pd
df = pd.read_csv('Iris.csv')
categorical_variable = 'Species'
categories = df[categorical_variable].unique()
categories
```

```python
for category in categories:
    # Select the subset of data for the current category
    subset = df[df[categorical_variable] == category]

    # Display the category name
    print(f"Category: {category}")

    # Calculate and display the statistical details
    print("Percentile:")
    print(subset.describe(percentiles=[0.25, 0.5, 0.75])['SepalLengthCm'])
    print(subset.describe(percentiles=[0.25, 0.5, 0.75])['SepalWidthCm'])
    print(subset.describe(percentiles=[0.25, 0.5, 0.75])['PetalLengthCm'])
    print(subset.describe(percentiles=[0.25, 0.5, 0.75])['PetalWidthCm'])

    print("Mean:")
    print(subset.mean()['SepalLengthCm'])
    print(subset.mean()['SepalWidthCm'])
    print(subset.mean()['PetalLengthCm'])
    print(subset.mean()['PetalWidthCm'])

    print("Standard Deviation:")
    print(subset.std()['SepalLengthCm'])
    print(subset.std()['SepalWidthCm'])
    print(subset.std()['PetalLengthCm'])
    print(subset.std()['PetalWidthCm'])

    print("-------------------------------------")
```
